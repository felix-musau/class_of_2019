```python
def filter_outlier (data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = 50
    upper_bound = 95
    filtered_data = []
    for x in data:
        if lower_bound <= x <= upper_bound:
            filtered_data.append(x)
    return filtered_data
new_math = filter_outlier(df['MATH'])
new_math
```

As we can see the code above, the function automatically filters the data thas exceeds 95% and below 50%. This is a great way to filter out outliers.

```python
#Recalculating the mean. It should change because the outliers ave been removed
new_math_mean = get_mean(new_math)
new_math_mean
```

As we can see the new mean has increased by almost 4 from the previous mean. This is how much impact outliers have on data.
